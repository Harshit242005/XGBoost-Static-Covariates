Reseting the dates

```python
def create_group_data(df):
    new_rows = []

    latest_date = pd.Timestamp('2025-03-31')

    # Group by HEAD_1_DESC
    for _, group in df.groupby('HEAD_1_DESC'):
        n = len(group)
        # Generate n dates: last day of each previous month
        dates = pd.date_range(end=latest_date, periods=n, freq='M')
        
        # Assign new dates
        group = group.copy()
        group['Report_Date'] = dates[::-1]  # reverse to go backward in time
        new_rows.append(group)

    # Combine all groups back into a DataFrame
    new_df = pd.concat(new_rows).reset_index(drop=True)
    return new_df
```

```python
branch1_updates = create_group_data(branch1)
branch2_updates = create_group_data(branch2)
branch3_updates = create_group_data(branch3)
branch4_updates = create_group_data(branch4)
branch5_updates = create_group_data(branch5)
```

```python
import pandas as pd
import numpy as np

# Step 1: Filter by branch
filtered_main_data = main_data[main_data['Branch Name '] == 'CO - Office Nariman Point'].copy()

# Step 2: Parse date and group
filtered_main_data['Report_Date'] = pd.to_datetime(filtered_main_data['Report_Date'])
grouped = filtered_main_data.groupby('Category')

# Step 3: Determine full date range
full_date_range = pd.date_range(
    start=filtered_main_data['Report_Date'].min(),
    end=filtered_main_data['Report_Date'].max(),
    freq='M'
)

# Step 4: Rebuild each group with full date range
resampled_groups = []
for cat, group in grouped:
    group = group.set_index('Report_Date')
    group = group.reindex(full_date_range)

    # Forward fill static columns
    for col in ['Branch Name ', 'Category', ' Branch Code ', 'Cluster ', 'Zone', 'SCHEDULE']:
        group[col] = group[col].fillna(method='ffill').fillna(method='bfill')

    # Fill NET BAL with small noise if missing
    if 'NET BAL' in group.columns:
        group['NET BAL'] = group['NET BAL'].interpolate(method='linear').fillna(method='bfill')
        noise = np.random.normal(scale=0.02, size=len(group))
        group['NET BAL'] *= (1 + noise)

    group = group.reset_index().rename(columns={'index': 'Report_Date'})
    resampled_groups.append(group)

# Step 5: Combine all
synthetic_balanced_data = pd.concat(resampled_groups, ignore_index=True)
synthetic_balanced_data
```

Pipeline creation for the training and inference data

```python
from darts.dataprocessing.transformers import StaticCovariatesTransformer
transformer_instance = StaticCovariatesTransformer()
transformed_multi_series = transformer_instance.fit_transform(multi_seires)
```
